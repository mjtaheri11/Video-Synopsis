```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw
```

read images

```
im_src = cv2.imread('drive/My Drive/HW4/1-1.JPG')
im2_dst = cv2.imread('drive/My Drive/HW4/1-2.JPG')
```

find corners of musaic


```
im = Image.open('drive/My Drive/HW4/1-1.JPG')
d = ImageDraw.Draw(im)

top_right = (1800, 850)
top_left = (1100, 750)
down_right = (1800, 1350)
down_left = (630, 1115)
line_color = (0, 0, 255)

d.line([top_right, top_left, down_left, down_right, top_right],
       fill=line_color, width=2)
im.save("drawn_grid.png")
```

```
pts_src = np.array([top_right, top_left, down_right, down_left])
pts_dst = np.array([[1000, 700],[1700, 700],[1000, 1400],[1700, 1400]])
h, status = cv2.findHomography(pts_src, pts_dst)
im_out = cv2.warpPerspective(im_src, h, (im_src.shape[1],im_src.shape[0]))
plt.imshow( im_out)
```

```
im2 = Image.open('drive/My Drive/HW4/1-2.JPG')
d = ImageDraw.Draw(im2)

top_right = (1080, 790)
top_left = (600, 850)
down_right = (1380, 1150)
down_left = (660, 1300)
line_color = (0, 0, 255)

d.line([top_right, top_left, down_left, down_right, top_right],
       fill=line_color, width=2)
im2.save("drawn_grid2.png")
```
